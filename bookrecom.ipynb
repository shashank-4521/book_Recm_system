popularity based recommendation system

we will display the books with highest average rating but we will consider only those books we got minimum 50 votes 

```python
# merget rating and books dataset based on isbn
ratings.merge(books,on='ISBN').shape
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head(2)
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].head(50)
```

```python
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

collabrative filtering based Rcommendation System

```python
# filtering based on users 
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
# boolean indexing 
x[x]
# those who rate 200+ books have true value 
```

```python
rated_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(rated_users)]
```

```python
# filtering based on books 
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # fetching index 
    index=np.where(pt.index==book_name)[0][0]
    # fetching the value of the index from the dataframe
    similar_item=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    data=[]
    for i in similar_item:

        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
    
# enumerate - key value for 
# list convrt it into a list 
# sort
# similarity_scores[0]-only the firsy item 
# [1:6] - only the first 5 items
# key we want the sorting not on the basis of index 
```

```python
recommend('Message in a Bottle')
```
